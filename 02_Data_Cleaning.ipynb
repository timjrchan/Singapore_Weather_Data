```python
# import libraries

import os
import chardet
import pandas as pd
import numpy as np
import plotly.express as px
from datetime import datetime

# set dataframe display 
pd.set_option('display.max_columns', None)
pd.set_option('display.max_colwidth', 10000)
pd.set_option('display.max_rows', None)
pd.set_option('display.width', 10000)
```

```python
# Rename files

SAVED_CHANGI_FILES = 'changi_csv_files'

changi_files = os.listdir(SAVED_CHANGI_FILES)

changi_files = sorted(changi_files)

month_map = {"jan": "01",
             "feb": "02",
             "mar": "03",
             "apr": "04",
             "may": "05",
             "jun": "06",
             "jul": "07",
             "aug": "08",
             "sep": "09",
             "oct": "10",
             "nov": "11",
             "dec": "12"}



for file in changi_files:  

    try:

        if file == ".DS_Store":
            continue
        else:
            # print(f"Before renaming file name: {file}") # debug

            year, month, locExtension = file.split("_")
            # print(year, month, locExtension)

            if month in month_map:
                renamed_file = f"{year}_{month_map[month]}_{month}_{locExtension}"
                # print(f"After renaming file: {renamed_file}") # debug
 

                old_file = os.path.join(SAVED_CHANGI_FILES, file)
                # print(f"Old file path: {old_file}") # debug


                new_file = os.path.join(SAVED_CHANGI_FILES, renamed_file)
                # print(f"New file path: {new_file}\n") # debug

                os.rename(old_file, new_file)

                print(f"File has been renamed from {old_file} to {new_file}")
    
    except Exception as e:
        print(e)
```

```python
# check for encoding

encoding_list = []

for file in os.listdir(SAVED_CHANGI_FILES):
    file_path = os.path.join(SAVED_CHANGI_FILES, file)

    with open(file_path, "rb") as f:
        data = f.read()

    encoding_result = chardet.detect(data)
    
    encoding = encoding_result["encoding"]

    encoding_list.append(encoding)

print(set(encoding_list))
```

```python
print(os.listdir(SAVED_CHANGI_FILES))
```

```python
# Convert the files into UTF-8 encoding

OUTPUT_FOLDER = "converted_changi_files"

os.makedirs(OUTPUT_FOLDER, exist_ok = True)

for file in os.listdir(SAVED_CHANGI_FILES):

    if file.endswith(".csv"):

        input_path = os.path.join(SAVED_CHANGI_FILES, file)
        output_path = os.path.join(OUTPUT_FOLDER, file)


        with open(input_path, "rb") as f:
            data = f.read()

        encoding_result = chardet.detect(data)
        
        original_encoding = encoding_result["encoding"]

        # print(f"{file} has {original_encoding} encoding.\nConverting to UTF-8...")

        try:
            with open(input_path, "r", encoding=original_encoding) as in_file:
                content = in_file.read()

            with open(output_path, "w", encoding = "UTF-8") as out_file:
                content = out_file.write(content)

            # print(f"File {file} converted to UTF-8 successfully\n")
        
        except Exception as e:
            print(e)
```

```python
# Rename each csv file to new headers

sample_df = pd.read_csv("converted_changi_files/1980_01_jan_changi.csv")

column_headers_map = {"Station": "station",
                      "Year": "year",
                      "Month": "month",
                      "Day": "date",
                      "Daily Rainfall Total (mm)": "tot_daily_rf",
                      "Highest 30 Min Rainfall (mm)": "highest_30_min_rf",
                      "Highest 30 min Rainfall (mm)": "highest_30_min_rf", # Formatting difference from 2020
                      "Highest 60 Min Rainfall (mm)": "highest_60_min_rf",
                      "Highest 60 min Rainfall (mm)": "highest_60_min_rf", # Formatting difference from 2020
                      "Highest 120 Min Rainfall (mm)": "highest_120_min_rf",
                      "Highest 120 min Rainfall (mm)": "highest_120_min_rf", # Formatting difference from 2020
                      "Mean Temperature (°C)": "mean_temp",
                      "Maximum Temperature (°C)": "max_temp",
                      "Minimum Temperature (°C)": "min_temp",
                      "Mean Wind Speed (km/h)": "mean_wind_speed",
                      "Max Wind Speed (km/h)": "max_wind_speed"
                      }

print(list(sample_df.columns.values))
```

### Append all csv into one DF

Visually looking though the headers, the headers are identical and in correct order. We will proceed with compiling all the csv files data into one dataframe

```python
# WORKINGS

# rename the column headers and concat the files into one DF

CONVERTED_FOLDER = "converted_changi_files"

# initialize empty list to store individual df
dataframe = []

# sort the fiels in the folder to order
sorted_folder = sorted(os.listdir(CONVERTED_FOLDER))

# loop the files to rename the headers and append into the list
for file in sorted_folder:

    file_path = os.path.join(CONVERTED_FOLDER, file)

    single_df = pd.read_csv(file_path)

    single_df = single_df.rename(columns=column_headers_map)

    dataframe.append(single_df)

# concat into single dataframe
merged_df = pd.concat(dataframe, ignore_index=True)
```

### Inspecting column values

Here, we check if the data columns are what we expect, numerical values, and objects. If any of the values are not expected, we will clean the column while retaining the values. In the bottom cell, we can see that the values in the **highest 120 min rainfall** has a mixed datatypes of numerical and objects; even white spaces within the object. We will clean the values up before continuing.

We have gathered that there are unicode differences between the uses of "-" and we will use regex to find all "-" and standardized into one "-" so we can replace it as NaN values. 

```python
# inspect datatypes
print(merged_df.dtypes)
```

```python
# convert year, month, date to string

# convert year to string
merged_df['year'] = merged_df['year'].astype(str)

# convert month to string
merged_df['month'] = merged_df['month'].astype(str)

# convert date to string
merged_df['date'] = merged_df['date'].astype(str)
```

```python
# inspecting column values

print(merged_df['highest_120_min_rf'].unique())
```

```python
# Cleaning of data fields 

column_headers = merged_df.columns.values 

for column in column_headers[4:]:

    # retain numerical values else strip whitespaces
    merged_df[column] = merged_df[column].apply(lambda x: x.strip() if isinstance(x, str) else x) 

    # unify all "-" into one "-"
    merged_df[column] = merged_df[column].replace(r"[‐‑‒–—―]", "-", regex = True)

    # replace "-" into np.nan values
    merged_df[column] = merged_df[column].replace("-", np.nan).infer_objects(copy=False)

    # convert the values into float
    merged_df[column] = merged_df[column].astype(float)
    
```

```python
# add in a new column that takes in year, month, and date into datetime object using vectorization

merged_df['compiled_date'] = pd.to_datetime(merged_df['year'] + "-" + merged_df['month'] + "-" + merged_df['date'],
                                            format = "%Y-%m-%d")

print(merged_df.head()) # inspect first 5 rows
```

```python
# reorder the column headers

ordered_column_headers = ["station", "compiled_date", "year", "month", "date", "mean_temp", "min_temp", "max_temp",
                          "tot_daily_rf", "highest_30_min_rf", "highest_60_min_rf", "highest_120_min_rf", 
                          "mean_wind_speed", "max_wind_speed"]


ordered_df = merged_df[ordered_column_headers]

print(ordered_df.head())
```
